```python
#PROGRAM 1 :

#Write a program to read marks from the keyboard 
#and your program should display equivalent grades according to following table.
#  Marks |   Grade 
# 100-80 | distinction
# 70-60  | first class
# 59-40  | second class 
# <40    | fail

m=int(input("enter marks : "))
if m<=100 and m>=80 :
  print("distiction")
elif m<=79 and m>=60 :
  print("first class")
elif m<=59 and m>=40 :
  print("second class")
else :
  print("fail")
```

```python
#PROGRAM 2 :

#Write a program to input basic salary of an employee and calculate gross salary according to given conditions :

#bs<=10000 then hra=20%,da=80% 
#20000>bs>=10001 then hra=20%,da=90%
#bs>=20000 then hra=30%,da=95%

bs=int(input("basic salary : "))
if bs<=10000 :
  hra=0.2*bs
  da=0.8*bs
elif bs<20000 and bs>=10001 :
  hra=0.2*bs
  da=0.9*bs
else :
  hra=0.3*bs
  da=0.95*bs
gs=bs+hra+da
print("gross salary : ",gs)
print("house rent allowance : ",hra)
print("dearness allowance : ",da)
```

```python
#PROGRAM 3 :

# Write a program for checking the speed of drivers.
# If speed is less than 70, it should print “Ok”. Otherwise, for every 5km above
# -the speed limit (70), it should give the driver one demerit point and print 
# -the total number of demerit points. For example, if the speed is 80, it 
# -should print: “Points: 2”. If the driver gets more than 12 points, the
# -function should print: “License suspended” .

s=int(input("enter speed : "))
if s<70 :
  print ("OK")
elif s>=70 and s<=130 :
  d=(s-70)/5 
  print("demerit points : ",d)
else :
  print("licence suspended")
```
