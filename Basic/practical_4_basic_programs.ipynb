```python
#PROGRAM 1 :

#Write a program to stimulate simple calculator by taking two operads as input and display the result.

num1=float(input('enter number 1 : '))
num2=float(input('enter number 2 : '))
print('addition : ', num1+num2)
print('subtraction : ', num1-num2)
print('multiplication : ', num1*num2)
print('division : ', num1/num2)
print('modular division : ', num1%num2)
print('floor division : ', num1//num2)
print('exponential : ', num1**num2)
```

```python
#PROGRAM 2 :

#Write a program to calculate the perimeter and area of geometric objects

#CIRCLE

radius=float(input('enter radius:'))
p=3.14
area=p*radius**2
circumference=2*p*radius
display(area,circumference)
```

```python
#TRIANGLE 

base=float(input('enter base:'))
height=float(input('enter height:'))
side1=float(input('enter side1:'))
side2=float(input('enter side2:'))
perimeter=side1+side2+base
area=0.5*base*height
display(perimeter,area)
```

```python
#RECTANGLE

side1=float(input('enter side1:'))
side2=float(input('enter side2:'))
area=side1*side2
perimeter=2*side1+2*side2
display(area,perimeter)
```

```python
#PROGRAM 3 :

#Write a program to interchange value of two numbers.

x=int(input('enter value of x:'))
y=int(input('enter value of y:'))
temp=x
x=y
y=temp
print("x=",x)
print("y=",y)
```
