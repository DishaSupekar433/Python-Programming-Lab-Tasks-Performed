```python
#PROGRAM 1 :

#Write a Python program to concatenate following dictionaries to create a new one. 
#Sample Dictionary : dic1={1:10, 2:20} dic2={3:30, 4:40} dic3={5:50,6:60}
#Expected Result : {1: 10, 2: 20, 3: 30, 4: 40, 5: 50, 6: 60}

dic1 = {1:10,2:20}
dic2 = {3:30,4:40}
dic3 = {5:50,6:60}

dic4 = {}
for d in (dic1, dic2, dic3): dic4.update(d)
print(dic4)
```

```python
#PROGRAM 2 : 

#Write a Python program to check whether a given key already exists in a dictionary.

d = {1:10, 2:20, 3:30, 4:40, 5:50, 6:60}
def iskey_present(x):
  if x in d :
    print('key is present in the dictionary')
  else :
    print('key is not present in the dictionary')
iskey_present(5)
iskey_present(10)
```

```python
#PROGRAM 3 : 

#Write a Python script to generate and print a dictionary that contains a number 
#(between 1 and n) in the form (x, x*x)
#Sample Dictionary ( n = 5) :
#Expected Output : {1: 1, 2: 4, 3: 9, 4: 16, 5: 25}

n = int(input("input a number : "))
d = dict()

for x in range (1,n+1) :
  d[x] = x*x

print(d)
```

```python
#PROGRAM 4 : 

#Write a Python program to merge two Python dictionaries.

d1 = {'a':10, 'b':200}
d2 = {'x':30, 'y':200}
d = d1.copy()
d.update(d2)
print(d)
```

```python
#PROGRAM 5 : 

#Write a Python program to get the maximum and minimum value in a dictionary.

my_dict = {'x':78597, 'y':89708, 'z':67856}
key_max=max(my_dict.keys(), key=(lambda k: my_dict[k]))
key_min=min(my_dict.keys(), key=(lambda k: my_dict[k]))

print('maximum value: ',my_dict[key_max])
print('minimum value: ',my_dict[key_min])
```

```python
#PROGRAM 6 : 

#Write a Python program to create set difference, Union and intersection.

setc1 = set(["green","blue"])
setc2 = set(["blue","yellow"])
r1 = setc1.difference(setc2)
print(r1)
setc1 = set (["green","blue"])
setc2 = set (["blue","yellow"])
print(setc1.union(setc2))
setc1 = set(["green","blue"])
setc2 = set(["blue","yellow"])
print(setc1.intersection(setc2))
```
