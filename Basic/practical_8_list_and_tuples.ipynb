```python
#PROGRAM 1 :

#Write a Python program to sum all the items in a list.

def sum_list(items):
  sum = 0
  for x in items :
    sum += x
  return sum
print(sum_list([5,8,-10]))
```

```python
#PROGRAM 2 :

#Write a Python program to multiply all the items in a list.

def multiplylist(mylist):
  result = 1
  for x in mylist :
    result = result * x
  return result 
list1 = [1, 2, 3]
list2 = [3, 2, 4]
print(multiplylist(list1))
print(multiplylist(list2))
```

```python
#PROGRAM 3 :

#Write a Python program to get the largest number from a list.

def max_num_in_list(list):
  max = list[0]
  for a in list :
    if a > max :
      max = a 
  return max 
print(max_num_in_list([1, 2, -8, 0]))
```

```python
#PROGRAM 4 :

#Write a Python program to get the smallest number from a list.

def smallest_num_in_list(list):
  min = list[0]
  for a in list :
    if a < min :
      min = a 
  return min
print(smallest_num_in_list([1, 2, -8, 0]))
```

```python
#PROGRAM 5 :

#Write a Python program to count all elements in list and count Occurrences Of A List Item In Python.

def count_occurrences(nums):
  result = dict (map(lambda el : (el, list(nums).count(el)),nums))
  return result
nums = [3,4,8,0,3,1,5,2,4,2]
print("original list : ")
print(nums)
print('\n elements in list : ', len(nums))
print("\ncount the occurences of the items in the said list : ")
print(count_occurrences(nums))
```

```python
#PROGRAM 9 :

#Write a Python program calculate the product of all the numbers given in tuple. 
#Original Tuple: (2, 4, 8, 8, 3, 2, 9)
#Product - multiplying all the numbers of the said tuple: 27648

def multiple_tuple(nums):
  temp = list(nums)
  product = 1
  for x in temp :
    product *= x
  return product 

nums = (2, 4, 8, 3, 2, 9, 8)
print("original tuple : ", nums)
print("product of all the numbers given in the tuple : ", multiple_tuple(nums))
```
