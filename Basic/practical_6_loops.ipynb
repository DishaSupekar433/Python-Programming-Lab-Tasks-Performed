```python
#PROGRAM 1 :

#Write a program to find the factorial of a number.

n = int(input("enter the number : "))
sum = 1
if n==0 or n==1 :
  print("factorial is 1")
else :
  for i in range (1,n,1) :
    sum = sum*(n-i)+sum
  print(sum)
```

```python
#PROGRAM 2 :

#Write a program to check whether the given number is palindrome or not.

num=int(input("enter number : "))
temp=num
rev=0
while(num>0):
  rem=num%10
  rev=rev*10+rem
  num=num//10
if(temp==rev):
    print("this value is palindrome")
else:
    print("not palindrome number")
```

```python
#PROGRAM 3 :

#Write a program to check whether the given number is armstrong or not.

num=int(input("enter number : "))
temp=num
rev=0
while(num>0) :
  rem=num%10
  rev=rev+rem**3
  num=num//10
print(rev)
if temp==rev :
  print(temp,"is an armstrong number")
else :
  print(temp,"not an armstrong number")
```

```python
#PROGRAM 4 :

#Write a program to print fibonacci series.

num=int(input('enter number : '))
count=0
a=0
b=1
while count < num :
  print(a)
  c=a+b
  a=b
  b=c
  count+=1
```

```python
#PROGRAM 5 :

#Write a python program which iterates the integers from 1-15, for multiple of 3, print "fizz" instead of the number
#and for multiple of 5 print "buzz", for numbers which are multiple of both 3 and 5, print "fizzbuzz".

for i in range (0,50,1):
  if(i%3==0 and i%5==0):
    print("fizzbuzz")
  elif(i%3==0):
    print("fizz")
  elif(i%5==0):
    print("buzz")
  else:
    print(i)
```

```python
#PROGRAM 6 :

#Write a program to check whether given number is prime number or not.

n=int(input("enter the number = "))
for i in range (2,n):
  if (n%i)==0:
    print(n,"is not a prime number")
    break
else:
  print(n,"is a prime number")
```

```python
#PROGRAM 7 :

#Write a program to check whether given number is perfect number or not.

n=int(input("enter the number = "))
sum1=0
for i in range (1,n):
  if(n%i==0):
    sum1=sum1+i
if(sum1==n):
      print("the number is perfect")
else:
      print("the number is not perfect")
```

```python
#PROGRAM 8 :

#Write a program to read a number and display the numbers in the reverse order, display the sum of the digits. 

num=int(input("enter any number : "))
rev=0
while(num > 0):
    a = num % 10
    rev = rev * 10 + a
    num = num // 10
 
print("reverse of the number is : ",rev)
```

```python
#PROGRAM 9 :

#Write a program to print the following patterns.

# I]
for i in range (1,6) :
  for j in range (1,1+i) :
    print ("*",end=" ")
  print()
```

```python
# II]
for i in range (6,0,-1) :
  for j in range (1,1+i) :
    print ("*",end=" ")
  print()
```

```python
# III]
for i in range (6) :
  for j in range (i) :
    print(j+1,end=" ")
  print()
```

```python
#IV]
for i in range (5,0,-1) :
  for j in range (i) :
    print (j+1,end=" ")
  print()
```
