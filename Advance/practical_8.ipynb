Exceptional handling in python programming

```python
try:
  a=5
  b=3
  print(a/b)
except:
  print("Some error occured")
print("Out of except block ")  
```

```python
try:
  a=5
  b=0
  print(a/b)
except ZeroDivisionError :
  print("Some error occured ")
print("Out of except block")
```

```python
try:
  a=5
  b='c'
  print(a/b)
except TypeError:
  print("Some error occured ")
print("Out of except block")
```

```python
try:
  a=5
  b="c"
  print(a/b)
except TypeError:
  print("Supported operation")
except ZeroDivisionError:
    print("Division by zero is not allowed")
print("Out of try and except block")
```

#Syntax 
  #try:

  #except :
    #execution when error in try block
  #else:
    #execution when try block is error free
  #finally:
    #executed irrespective of exception occured or not

```python
try:
  print("Try block")
  x=int(input("Enter any number : "))
  y=int(input("Enter anynumber : "))
  z=x/y
except ZeroDivisionError:
  print("Except Zero Division Error Block")
  print("Division by zero not excepted ")
else:
  print("Else Block")
  print("Division : ",z)
finally:
  print("Finally block")
  x=0
  y-0
print("Out of try, except, else and finally block")
```

```python
#Read two numbers n1 and n2. Write a function to compute n1/n2 and use try/except to catch the exceptions.

try:
  print("Try block")
  n1=int(input("Enter n1 : "))
  n2=int(input("Enter n2 : ")) 
  n3=input("Enter n3 : ")
  n4=int(input("Enter n4 : "))
  div=n1/n2
  add=n3+n4
except ZeroDivisionError:
  print("Except Zero Division Error Block")
  print("Division by zero not excepted ")
except ValueError:
  print("Character not allowed")
else:
  print("Else Block")
  print("Division : ",div)
  print("Addition : ",add)
finally:
  print("Finally block")
  n1=0
  n2=0
print("Out of try, except, else and finally block")
```

```python
#Read two numbers n1 and n2. Write a function to compute n1/n2 and use try/except to catch the exceptions.

def compute (n1,n2):
  try : 
    div = n1/n2
    add = n1+ n2
    print(n1,"/",n2,"=",div)
    print(n1,"+",n2,"=",add)
  except ZeroDivisionError: 
    print(n1,"/",n2,"= Sorry! you are dividing by 0")
  except TypeError:
    print(n1,"+",n2,"Sorry! Character not allowed")

compute(20,2)
compute(3,0)
compute("a",3)
```
