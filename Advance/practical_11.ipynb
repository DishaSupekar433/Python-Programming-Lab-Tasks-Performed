```python
import sqlite3
connection = sqlite3.connect("gfg.db")
crsr = connection.cursor()
print("Connected to the database")
crsr.execute("DROP TABLE IF EXISTS emp")
sql_command = '''CREATE TABLE emp (staff_number INTEGER PRIMARY KEY,fname VARCHAR(20),lname VARCHAR(30),gender CHAR(1),joining DATE);'''
crsr.execute(sql_command)
crsr.execute( '''INSERT INTO emp VALUES (23, "Rish",\"Bansal", "M", "2014-03-28")''')
crsr.execute( '''INSERT INTO emp VALUES (24, "Meridth",\"Grey", "F", "2003-03-04")''')
crsr.execute('''INSERT INTO emp VALUES (1, "Bill", "Gates",\"M", "1980-10-28")''')
connection.commit()
crsr.execute('''SELECT * FROM emp''')
ans = crsr.fetchall()
connection.close()
for i in ans:
  print(i)
```
