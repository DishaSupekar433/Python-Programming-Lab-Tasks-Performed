```python
# WAP using while loop to display the value of rating of an album playlist stored in the list playlist rating 
# if the score is less than 6 then exit the loop the playlist rating is given by
# playlist_rating = [10,9.5,10,8.5,7.5,5,10,10]

PlayListRatings = [10,9.5,10,8,7.5,5,10,10]
i = 0
while(not (PlayListRatings[i] < 6)):
    print(PlayListRatings[i])
    i += 1
```
