```python
#write a program to count the number of characters in the string and store them in a dictionary data structure.

s=input("enter a string : ")
d={}
for i in s :
  if i in d:
    d[i] += 1
  else:
    d[i] = 1
print(d)
```

```python
s=input("enter a string : ")
d={}
for i in s :
  d[i]=s.count(i)
print(d)
```
