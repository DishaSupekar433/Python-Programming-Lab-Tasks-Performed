```python
#2. Write a program to perform various operations on matrices.

import numpy as np
A = np.array([[2,1,-3],[-3,-6,1],[1,1,1]])
B = np.array([[3,1,5],[-6,3,4],[1,8,2]])
print("A=\n",A)
print("B=\n",B)
add = np.add(A,B)
print("Addition=\n",add)
sub1 = np.subtract(A,B)
print("Subtraction1=\n",sub1)
sub2 = np.subtract(B,A)
print("Subtraction2=\n",sub2)
mul = np.multiply(A,B)
print("Multiplication=\n",mul)
div = np.divide(A,B)
print("Division=\n",div)
mod = np.mod(A,B)
print("Modulus=\n",mod)
transpose_A = np.transpose(A)
print("Transpose_A=\n",transpose_A)
transpose_B = np.transpose(B)
print("Transpose_B=\n",transpose_B)
```

```python
#6. Create a class ATM and define ATM operations to create account, deposit, check_balance, withdraw and delete account. 
#Use constructor to initialize members.

class bankaccount :
  
  def __init__(self):
    self.balance=0

  def deposit(self):
    a=(int(input("Enter the amount to be deposited : ")))
    self.balance+=a
    print("Amount deposited : ",a)

  def withdraw(self):
    x=int(input("How much amount you want to withdraw ? :"))
    if self.balance<x:
      print("Insufficient balance in account")
    else :
      self.balance=self.balance-x
      print("Final balance = ",self.balance)

balance=bankaccount()
balance.deposit()
balance.withdraw()
```

```python
#7. Make a class Employee with a name and salary. Make a class Manager inherit from Employee. 
#Add an instance variable, named department. Write a method that prints manager's name, department and salary. 
#Make a class Executive inherit from Manager.
#Write a method that prints the string "Executive" followed by the information stored in the Manager super class object.

class Employee :
  def __init__(self,name,salary) :
    self.name = name
    self.salary = salary
class Manager(Employee) : 
  def __init__(self,department) :
    self.department = department

  def print_info(self) :
    print("Name : ",self.name)
    print("Salary : ",self.salary)
    print("Department : ",self.department)

class Executive(Manager) : 
  def __init__(self) :
    print("Executive")
    
obj=Executive()
obj.name = input("Enter name of the employee : ")
obj.salary=int(input("Enter salary of the employee : "))
obj.department=input("Enter department : ")
obj.print_info()
```

```python
#8. Read two numbers n1 and n2. Write a function to compute n1/n2 and use try/except to catch the exceptions.

def compute (n1,n2):
  try : 
    div = n1/n2
    print(n1,"/",n2,"=",div)
    
  except ZeroDivisionError: 
    print(n1,"/",n2,"= Sorry! you are dividing by 0")
  except TypeError:
    print(n1,"/",n2,"Sorry! Character not allowed")

compute(20,2)
compute(3,0)
compute("a",3)
```

```python
#9. Write a Python program to detect and handle the exception while solving the quadratic equation.

from math import sqrt
print ("Quadratic equation is in the form :  a*x**2 + b*x + c = 0.")

done = False

while not done : 
  try:
    a = int(input("Enter value of a : "))
    b = float(input("Enter value of b : "))
    c = float(input("Enter value of c : "))
    d = b**2 - 4*a*c
    sq = sqrt(d)
    root1 = (-b - sq)/(2*a)
    root2 = (-b + sq)/(2*a)
    print("First root : ", root1)
    print("Second root : ", root2)
    done = True 

  except ValueError : 
    print("Value error, wrong equation. Try again.")

  except ZeroDivisionError :
    print("Zero division error, 'a' can't be 0. Try again.")
```
