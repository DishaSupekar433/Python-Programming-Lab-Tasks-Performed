```python
#mean

L = [1,2,3,4,5,6,7,8]
count=0
sum=0
for i in L:
  count = count + 1
  sum = sum+i
avg=sum/count
print("number of elements in list : ",count)
print("sum of elements in list : ",sum)
print("mean of elements in list : ",avg)
```

```python
#median 

L = [4,2,3,4,5,4,7,8]
n=0
for i in L:
  n = n + 1
L.sort()
  
if n % 2 == 0:
    median1 = L[n//2]
    median2 = L[n//2 - 1]
    median = (median1 + median2)/2
else:
    median = L[n//2]
print("Median is : " ,median)
```

```python
#mode

def mode(list):
  i=0
  for i in range (0,len(list)):
    a=list.count(i)
    if a>=2:
      print(i)
list = [4,2,3,4,5,4,7,8]
mode(list)
```

```python
#with using function
import statistics as s
L = [1,2,3,4,5,6,7,8,4,4,4]
x=s.mean(L)
y=s.median(L)
z=s.mode(L)
print("mean:",x)
print("median:",y)
print("mode:",z)
```
